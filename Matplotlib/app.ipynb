Markers
You can use the keyword argument marker to emphasize each point with a specified marker:



```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,marker='o')
plt.show()
```

Format Strings fmt
You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color


```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,'o:b')
plt.show()
```

```python
# MARKER SIZE 
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints,marker='o',ms=20,mec = 'hotpink', mfc = 'hotpink')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints , linestyle="dashed")
plt.show()
```

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1={"family":"serif","color":'blue',"size":20}
font2={"family":"serif","color":'darkred',"size":15}
plt.plot(x,y)
plt.title("Sports Watch Data",fontdict=font1,loc="right")  #Title 
plt.xlabel("Average Pulse",fontdict= font2)   #X-axis Values 
plt.ylabel("Calorie Burnage",fontdict= font2)  #Y axis values 
plt.show()
```

Add Grid Lines to a Plot
With Pyplot, you can use the grid() function to add grid lines to the plot.



```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()
plt.show()
```

## Multiple Plots
## With the subplot() function you can draw multiple plots in one figure:




```python
import matplotlib.pyplot as plt
import numpy as np

# Dummy data for all plots
x = np.array([0, 1, 2, 3, 4])

y1 = np.array([1, 2, 3, 4, 5])
y2 = np.array([5, 4, 3, 2, 1])
y3 = np.array([2, 3, 2, 3, 2])
y4 = np.array([1, 4, 9, 16, 25])

# 1st plot (top-left)
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='b')
plt.title("Plot 1")

# 2nd plot (top-right)
plt.subplot(2, 2, 2)
plt.plot(x, y2, color='r')
plt.title("Plot 2")

# 3rd plot (bottom-left)
plt.subplot(2, 2, 3)
plt.plot(x, y3, color='g')
plt.title("Plot 3")

# 4th plot (bottom-right)
plt.subplot(2, 2, 4)
plt.plot(x, y4, color='m')
plt.title("Plot 4")

# Adjust layout so titles don't overlap
plt.tight_layout()

# Show all plots
plt.show()
```
