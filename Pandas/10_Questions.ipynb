```python
# Q 1: Create a Pandas Series from a Python list and perform
# basic operations like filtering, sorting, etc.

data=[40,10,20,50,30,60,70]
series=pd.Series(data)
series_filtered=series[series>30]
series_sorted=series.sort_values()
print("Original Series:\n",series)
print("Filtered Series (values > 30):\n",series_filtered)
print("Sorted Series:\n",series_sorted)
```
