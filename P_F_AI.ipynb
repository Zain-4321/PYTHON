```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Part A â€“ Descriptive Statistics

# 1. Load the dataset
df = pd.read_csv('passenger_data.csv')  # Make sure this file is in the same directory

# 2. Calculate descriptive statistics
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nMode:\n", df.mode(numeric_only=True).iloc[0])
print("\nVariance:\n", df.var(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
print("\nSkewness:\n", df.skew(numeric_only=True))
print("\nKurtosis:\n", df.kurtosis(numeric_only=True))

```

```python
print(df.describe())
```

```python

sns.histplot(data=df, x='age')
plt.title("Histogram of Age")
plt.show()
```

```python
sns.boxplot(data=df, x='age')
```

```python
print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))
```

```python
# 2. T-test example: Compare Age of males and females
# Make sure 'Sex' and 'Age' exist in your dataset
male_age = df[df['gender'] == 'M']['age'].dropna()
female_age = df[df['gender'] == 'F']['age'].dropna()

t_stat, p_val = ttest_ind(male_age, female_age, equal_var=False)
print(f"\nT-test (Age: Male vs Female):\nT-statistic = {t_stat}, P-value = {p_val}")
```
