2. Data Cleaning
1. Fill missing values in the "age" column with the median age.
2. Drop the "cabin" column (since it has many missing values).
3. Replace missing values in "embark_town" with the most frequent value.


```python
df["Age"].fillna(df['Age'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
# df["Embarked"].mode()
# df["Embarked"].fillna(df["Embarked"].mode()[0],inplace=True)
df["Embarked"].mode()
df["Embarked"].fillna(df['Embarked'].mode()[0],inplace=True)
```

4. Aggregation & Grouping
1. Find the average fare paid by each class.
2. Count the number of passengers based on their embarkation town.
3. Find the maximum age of passengers in each gender category.


```python
avg_fare=df.groupby("Pclass")["Fare"].mean()
print(avg_fare)
```

```python
embarke_count=df["Embarked"].count()
print("Count by embarked toen ",embarke_count)
```

```python
each_gen=df.groupby("Sex")["Age"].max()
print("Maximum Age gender ",each_gen)
```

6. Data Visualization (Using Pandas or Seaborn)
1. Plot a bar chart showing the number of survivors by gender.
2. Create a histogram of the age distribution.
3. Generate a box plot to see the distribution of fare by passenger class.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
survivors_by_gendr=df[df["Survived"]==1]["Sex"].value_counts()
survivors_by_gendr.plot(kind="bar",color="skyblue")
plt.title("Number of Survivors by Gender'")
plt.xlabel("Gender")
plt.ylabel("Number Of Servivor")
plt.show()
```

```python
df['Age'].plot(kind='hist', color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df, palette='pastel')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```
