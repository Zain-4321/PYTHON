```python
#  Create a function that takes a list of numbers and returns
# the NumPy array sorted in ascending order
def sort_array(num_list):
    arr=np.array(num_list)
    return np.sort(arr)

print(sort_array([5,2,9,1,5,6]))  
```

```python
#  Given a list of lists, create a 2D NumPy array and find the
# sum of elements in each row and column
list_of_lists=[[1,2,3],[4,5,6],[7,8,9]]
arr=np.array(list_of_lists)
print("Sum Of Each Row",arr.sum(axis=1))
print("Sum of Each Column",arr.sum(axis=0))
```

```python
# Create a program that uses NumPy to calculate the
# inverse of a 2x2 matrix
import numpy as np


mat = np.array([[4, 7],
                [2, 6]])

# Calculate inverse
inv_mat = np.linalg.inv(mat)

print("Original Matrix:\n", mat)
print("\nInverse Matrix:\n", inv_mat)
```

```python
#  Implement a function that takes a NumPy array and
# returns the transpose of the array.

import numpy as np

def transpose_matrix(arr):
    return arr.T   

mat = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Matrix:\n", mat)
print("\nTranspose Matrix:\n", transpose_matrix(mat))
```
