### Task 2: Basic Array Operations

2. Calculate the standard deviation and variance for the dataset.
3. Normalize the data using Z-score normalization:
z=x−μσz = \frac{x - \mu}{\sigma}


```python
import numpy as np

# 1. 
data = np.genfromtxt('iris.csv', delimiter=',', skip_header=1)  

# 2 
means = np.mean(data, axis=0)
max_vals = np.max(data, axis=0)
min_vals = np.min(data, axis=0)

print("Column-wise Mean:", means)
print("Column-wise Max:", max_vals)
print("Column-wise Min:", min_vals)

# 2. 
standard = np.std(data, axis=0)  
variance = np.var(data, axis=0)  

print("Standard Deviation:", standard)
print("Variance:", variance)

# Z-score 
z_normalized = (data - means) / standard

print("Z-score normalized data (first 5 rows):\n", z_normalized[:5])
```

# BFS

```python
import collections

def bfs(graph, root, goal):
    visited = set()
    queue = collections.deque([root])
    visited.add(root)

    while queue:
        vertex = queue.popleft()
        if vertex == goal:
            print(f"Goal {goal} found!")
            return
        for i in graph[vertex]:
            if i not in visited:
                queue.append(i)
                visited.add(i)
    print(f"Goal {goal} not found.")

if __name__ == "__main__":
    graph = {0: [1, 2, 3], 1: [0, 2], 2: [0, 1, 4], 3: [0], 4: [2]}
    bfs(graph, 0, 4)
```

# DFS

```python


# Example graph (adjacency list format)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
visited = []

def dfs(graph, node):
    if node not in visited:
        print(node, end=' ')
        visited.append(node)
        for neighbor in graph[node]:
            dfs(graph, neighbor)


# DFS traversal starting from node 'A'
dfs(graph, 'A')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Set seaborn theme
sns.set(style="whitegrid")

# 1 Scatter Plot: total_bill vs tip with color by sex

plt.figure(figsize=(7, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette="Set1")
plt.title("Total Bill vs Tip (Colored by Sex)")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
# 2 Subplots: sine and cosine
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x, y_sin, label="Sine", color='blue')
axes[0].set_title("Sine Wave")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

axes[1].plot(x, y_cos, label="Cosine", color='green')
axes[1].set_title("Cosine Wave")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.show()

# 3 Bar Plot: Average total_bill for each day

plt.figure(figsize=(7, 5))
avg_bill = tips.groupby('day')['total_bill'].mean().reset_index()
sns.barplot(data=avg_bill, x="day", y="total_bill", palette="pastel")
plt.title("Average Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()

# 4 Histogram: Tips
plt.figure(figsize=(7, 5))
sns.histplot(tips['tip'], bins=10, kde=False, color='orange')
plt.title("Distribution of Tip Amounts")
plt.xlabel("Tip Amount ($)")
plt.ylabel("Frequency")
plt.show()
# 5 Boxplot: Total bill grouped by day
plt.figure(figsize=(7, 5))
sns.boxplot(data=tips, x="day", y="total_bill", palette="Set2")
plt.title("Boxplot of Total Bills by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
# 6 Pie Chart: Smoker vs Non-Smoker

smoker_counts = tips['smoker'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', colors=["#66b3ff", "#ff9999"])
plt.title("Smoker vs Non-Smoker")
plt.axis('equal')
plt.show()
```

```python
# 1 Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 2 Define the first-order ODE: dy/dt = -2y + 1
def ode_function(t, y):
    return -2*y + 1

# 3 Solve the ODE with initial condition y(0) = 0
# Time span: from t=0 to t=5
solution = solve_ivp(ode_function, t_span=(0, 5), y0=[0], t_eval=np.linspace(0, 5, 100))

# 4 Plot the result
plt.figure(figsize=(8, 5))
plt.plot(solution.t, solution.y[0], label="y(t)", color='blue')
plt.title("Solution of dy/dt = -2y + 1")
plt.xlabel("Time (t)")
plt.ylabel("y(t)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, skew, kurtosis
df = pd.read_csv("Titanic-Dataset.csv")  # Make sure the file is in the same directory
#  View the first few rows to confirm
print(df.head())
# Filter numerical columns
num_df = df[['Age', 'Fare']].dropna()
# 2 Calculate statistics
print(" Descriptive Stats for 'age' column")
print("Mean:", num_df['Age'].mean())
print("Median:", num_df['Age'].median())
print("Mode:", num_df['Age'].mode().values[0])
print("Variance:", num_df['Age'].var())
print("Standard Deviation:", num_df['Age'].std())
print("Skewness:", skew(num_df['Age']))
print("Kurtosis:", kurtosis(num_df['Age']))
# 3 Summary using describe()
print("\r df.describe():")
print(num_df.describe())
# 4 Distribution plots
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(num_df['Age'], bins=20, kde=True)
plt.title("Distribution of Age")
plt.subplot(1, 2, 2)
sns.boxplot(x=num_df['Age'])
plt.title("Boxplot of Age")
plt.tight_layout()
plt.show()
```

```python
# 1 Correlation matrix
print(" Correlation Matrix:")
print(num_df.corr())

# 2 t-test: compare average age of survivors vs non-survivors
group1 = df[df['Survived'] == 1]['Age'].dropna()
group2 = df[df['Survived'] == 0]['Age'].dropna()

t_stat, p_val = ttest_ind(group1, group2, equal_var=False)

# 3 Interpret results
print("\n T-Test Result:")
print("T-statistic:", t_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print("→ Significant difference in age between survivors and non-survivors.")
else:
    print("→ No significant difference in age between survivors and non-survivors.")
```
