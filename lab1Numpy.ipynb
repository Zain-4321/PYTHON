Task 2: Basic Array Operations
1. Compute the mean, max, and min for each column.
2. Calculate the standard deviation and variance for the dataset.
3. Normalize the data using Z-score normalization:
z=x−μσz = \frac{x - \mu}{\sigma}


```python

data=np.genfromtxt("iris.csv" ,delimiter="," ,skip_header=1)
#1
means=np.mean(data,axis=0)
max=np.max(data,axis=0)
min=np.min(data,axis=0)
print("Means of columns ",means)
print("max of cloumns ",max)
print("min of columns ",min)
#2
standard=np.std(data,axis=0)
variance=np.var(data,axis=0)
print("Standard Value ",standard)
print("Variance value ",variance)
#3
z_normilazid=(data-means)/standard
print("Normilized the data using Z",z_normilazid[:5])
```

Task 3: Indexing and Slicing
1. Extract only the Sepal Length column.
2. Get the values for the first 10 flowers.
3. Extract flowers where Petal Length > 1.5.


```python
import numpy as np 
data=np.genfromtxt("iris.csv" ,delimiter="," ,skip_header=1)
#1
sepal_length=data[:,0] #every row first column
print("Sepal lenght",sepal_length)
#2
first_10_flowers=data[:10,:]  #data[start_row:end_row, start_col:end_col] data[rows, columns]
print("First 10 flowers ",first_10_flowers)
#3
petal_length_column = data[:, 2]  
flowers_with_petal_length_gt_1_5 = data[petal_length_column > 1.5, :]

print("Extract flowers where Petal Length > 1.5.",flowers_with_petal_length_gt_1_5)
```

Task 4: Advanced Operations

2. Count how many flowers have Sepal Width greater than the mean.

```python
#1
import numpy as np
row1=data[0]
row2=data[2]
enclidean_distance=np.linalg.norm(row1-row2)
print("1. Find the Euclidean distance between the first two rows.",enclidean_distance)
```

Task 5: Array Reshaping and Stacking
1. Reshape the array to simulate batches of size 30.
2. Stack two feature columns horizontally.
3. Create a boolean mask to filter rows with Petal Width < 0.5.





```python
rows, cols = data.shape

# Step 1: Trim the data so it’s divisible by 30
trimmed_rows = (rows // 30) * 30  # e.g., if rows=745, trimmed_rows=720
trimmed_data = data[:trimmed_rows, :]  # remove extra rows

# Step 2: Reshape into batches of 30
batched_data = trimmed_data.reshape(-1, 30, cols)
print("Shape after batching:", batched_data.shape)
```

```python
#2
col1 = data[:, 0]  
col2 = data[:, 2]  

stacked = np.column_stack((col1, col2))
print("Horizontally stacked columns (SepalLength & PetalLength):\n", stacked)
```

```python
#3
petal_width = data[:, 3]  

mask = petal_width < 0.5
filtered_rows = data[mask]

print("Rows where Petal Width < 0.5:\n", filtered_rows)
```
